```python
import h5py
import numpy as np
np.random.seed(42)
```

```python
file_name = 'lung.h5'
with h5py.File(f'/public/home/zhangshikang/project/DATA/semi_data/{file_name}', 'r') as f:
    keys = list(f.keys())
    data_length = len(keys)
    keys = np.random.permutation(keys)
    train_keys = keys[:int(data_length * 0.8)]
    val_keys = keys[int(data_length * 0.8):]
    with h5py.File(f'/public/home/zhangshikang/project/DATA/semi_data/split/train/{file_name}', 'w') as f_train:
        for key in train_keys:
            f_train.create_group(key)
            for k in f[key].keys():
                f_train[key].create_dataset(k, data=f[key][k])
        # f_train.attrs['cell_list'] = f.attrs['cell_list']
        f_train.close()
    with h5py.File(f'/public/home/zhangshikang/project/DATA/semi_data/split/eval/{file_name}', 'w') as f_val:
        for key in val_keys:
            f_val.create_group(key)
            for k in f[key].keys():
                f_val[key].create_dataset(k, data=f[key][k])
        # f_val.attrs['cell_list'] = f.attrs['cell_list']
        f_val.close()
    f.close()
```
